# Step 2: Missing Value Analysis

Missing values can affect data quality and analytical results. Therefore, we identify all missing values present in the dataset and calculate their frequency for each column.

### Count Missing Values

```python
missing_values = df.isnull().sum()

print("Missing Values in Each Column")
print(missing_values)
```

### Missing Value Percentage

```python
missing_percentage = (df.isnull().sum() / len(df)) * 100

missing_summary = pd.DataFrame({
    'Missing Count': df.isnull().sum(),
    'Missing Percentage': missing_percentage
})

missing_summary
```

### Columns Containing Missing Values Only

```python
missing_summary[missing_summary['Missing Count'] > 0]
```

# Step 3: Handling Missing Values

The dataset contains missing values only in the CouponCode column. These missing values do not represent data entry errors; rather, they indicate that no coupon was applied during the transaction.

To preserve all transaction records and improve data consistency, missing values in the CouponCode column are replaced with "No Coupon".

```python
# Replace Missing Coupon Codes

df['CouponCode'] = df['CouponCode'].fillna('No Coupon')
```

```python
# Verify Missing Values After Treatment

df['CouponCode'].isnull().sum()
```

### Result

All missing values in the CouponCode column were successfully handled by replacing null entries with "No Coupon".

Remaining Missing Values: 0

# Step 4: Duplicate Record Detection

Duplicate records can lead to inaccurate analysis and misleading results. Therefore, the dataset is checked for both complete duplicate rows and duplicate Order IDs.

```python
# Check Complete Duplicate Rows
duplicate_rows = df.duplicated().sum()

print("Duplicate Rows:", duplicate_rows)
```

```python
# Check Duplicate Order IDs
duplicate_order_ids = df['OrderID'].duplicated().sum()

print("Duplicate Order IDs:", duplicate_order_ids)
```

# Step 5: Date Format Validation

Date fields must follow a consistent format for accurate analysis and reporting.

In this step, the Date column is converted into a proper datetime format and checked for invalid or incorrectly formatted values.

```python
# Check Current Data Type of Date Column

print(df['Date'].dtype)
```

```python
# Convert Date Column to Datetime Format

df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
```

```python
# Check Invalid Dates

invalid_dates = df['Date'].isnull().sum()

print("Invalid Dates:", invalid_dates)
```

### Result

All dates were successfully converted into datetime format.

Number of Invalid Dates Found: 0

Therefore, the dataset contains no incorrectly formatted date values.

# Step 7: Data Consistency Validation

To ensure data accuracy, the relationship between Quantity, UnitPrice, and TotalPrice is verified.

Expected Formula:

TotalPrice = Quantity × UnitPrice

Any mismatch would indicate a potential data quality issue.

```python
# Calculate Expected Total Price

df['CalculatedTotal'] = df['Quantity'] * df['UnitPrice']
```

```python
# Find Mismatched Records

mismatch_count = (
    round(df['CalculatedTotal'], 2) !=
    round(df['TotalPrice'], 2)
).sum()

print("Total Price Mismatches:", mismatch_count)
```

### Result

The TotalPrice values are consistent with the Quantity and UnitPrice values.

Number of Mismatched Records: 0

Therefore, no calculation errors were detected in the dataset.
